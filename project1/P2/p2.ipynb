```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img1 = cv2.imread("image/problem2_1.bmp",cv2.CV_LOAD_IMAGE_UNCHANGED)
img2 = cv2.imread("image/problem2_2.bmp",cv2.CV_LOAD_IMAGE_UNCHANGED)

height1, width1 = img1.shape[:2]
height2, width2 = img2.shape[:2]
```

```python
hist1 = np.zeros((256))
for i in range(height1):
    for j in range(width1):
        hist1[img1[i][j]] += 1

plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(hist1.size),hist1,color='b')
plt.savefig('2_1_1.png')
plt.show()
```

```python
a = np.array([210.0,255.0,255.0])
b = np.array([30.0,255.0,255.0])
h_map = np.zeros((256))
for i in range(256):
    if i<b[0]:
        h_map[i] = a[0]*i/b[0]
    else:
        h_map[i] = a[0] + (a[1]-a[0])*(i-b[0])/(b[1]-b[0])
        
plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(h_map.size),h_map,color='b')
plt.show()
```

```python
out1 = np.zeros((height1, width1))
for i in range(height1) :
    for j in range(width1) :
        out1[i][j] = h_map[img1[i][j]]

plt.subplot(111),plt.title('Output Image')
plt.imshow(out1,cmap = 'gray')
plt.savefig('2_1_2.png')
plt.show()        
```

```python
hist1 = np.zeros((256))
for i in range(height1):
    for j in range(width1):
        hist1[out1[i][j]] += 1
plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(hist1.size),hist1,color='b')
plt.savefig('2_1_3.png')
plt.show()
```

```python
hist2 = np.zeros((256))
for i in range(height2):
    for j in range(width2):
        hist2[img2[i][j]] += 1

plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(hist2.size),hist2,color='b')
plt.savefig('2_2_1.png')
plt.show()
```

```python
hist_sum = np.zeros((256))
for i in range(1,hist_sum.size):
    hist_sum[i] = hist2[i] + hist_sum[i-1]
hist_map = np.zeros((256))
for i in range(hist_map.size):
    hist_map[i] = hist_sum[i] / height2 / width2 * (hist_sum.size - 1)
    hist_map[i] = int(hist_map[i])
    
plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(hist_map.size),hist_map,color='b')
plt.show()
```

```python
out2 = np.zeros((height2, width2))
for i in range(height2) :
    for j in range(width2) :
        out2[i][j] = hist_map[img2[i][j]]

plt.subplot(111),plt.title('Output Image')
plt.imshow(out2,cmap = 'gray')
plt.savefig('2_2_2.png')
plt.show()
```

```python
hist2 = np.zeros((256))
for i in range(height2):
    for j in range(width2):
        hist2[out2[i][j]] += 1

plt.title("Histogram"),plt.xlabel('Input Gray level'),plt.ylabel('Output Gray level')
plt.plot(range(hist2.size),hist2,color='b')
plt.savefig('2_2_3.png')
plt.show()
```
