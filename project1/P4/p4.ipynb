```python
import cv2
import sys
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("image/problem4.bmp",cv2.CV_LOAD_IMAGE_UNCHANGED)
height, width = img.shape[:2]
resize = input("resize:")
```

```python
out = np.zeros((resize,resize))
for i in range(resize):
    for j in range(resize):
        x = int(i/((resize-1)/float(height-1)))
        y = int(j/((resize-1)/float(width-1)))
        if x >= 150:
            x = 149
        if y >= 150:
            y = 149
        out[i][j]=img[x][y]
```

```python
plt.subplot(111),plt.title("Out1")
plt.imshow(out,cmap = 'gray')
plt.show()
```
